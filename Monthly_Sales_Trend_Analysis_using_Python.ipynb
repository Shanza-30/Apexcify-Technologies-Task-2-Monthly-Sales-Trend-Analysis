**Step 1: Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Step 3: Prepare Monthly Data**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
```

```python
df['Month'] = df['Order Date'].dt.month
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()
```

**Step 4: Plot Line Chart**

```python
plt.plot(monthly_sales.index, monthly_sales.values)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")

plt.show()
```

**Step 5: Month Names Show**

```python
monthly_sales.index = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

**Step 6: Improve Graph Styling and label data**

```python
plt.figure(figsize=(10,5))

plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', linewidth=2)

plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend Analysis")

plt.grid()

for i, value in enumerate(monthly_sales.values):
    plt.text(i, value + 10000, round(value,2), ha='center')

plt.show()
```

**Step 7: Save the Graph**

```python
plt.savefig("monthly_sales.png")
plt.show()
```
